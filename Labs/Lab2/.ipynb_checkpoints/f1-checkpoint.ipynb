### Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.spatial.distance as spd
import pickle
```

## Parte 1

```python
fName = 'LAB2distancias001.p'
D = pickle.load(open(fName, 'rb'))
```

```python
print(D.keys())
x = D['X']
y = D['idxClass']

print(x.shape)
print(y.shape)
plt.plot(y) #verificar se os dados est√£o ordenados
```

```python
plt.plot(x[0,y==1], x[1,y==1],'.')
plt.plot(x[0,y==2], x[1,y==2],'.')
plt.plot(x[0,y==3], x[1,y==3],'.')

plt.axis('scaled')
plt.grid(True)
```

```python
D2 = spd.squareform(spd.pdist(x.T, 'euclidean'))
D1 = spd.squareform(spd.pdist(x.T, 'cityblock'))
Dc = spd.squareform(spd.pdist(x.T, 'cosine'))

plt.figure(figsize=(18,9))

plt.subplot(131)
plt.imshow(D2)
plt.axis('off')

plt.subplot(132)
plt.imshow(D1)
plt.axis('off')

plt.subplot(133)
plt.imshow(Dc)
plt.axis('off')
```

```python
nx = np.sqrt(np.sum(x**2 , axis=0))

xn = x/nx
Dc2 = 1-np.dot(xn.T, xn)

plt.figure(figsize=(18,9))

plt.subplot(131)
plt.plot(xn[0,:], xn[1,:],'.')
plt.axis('scaled')
plt.grid(True)

plt.subplot(132)
plt.imshow(Dc)
plt.axis('off')

plt.subplot(133)
plt.imshow(Dc2)
plt.axis('off')
```

```python
print(Dc[:3,:3]) # comparar os calculos com usando o scikit learn e metodos nativos python
print(Dc2[:3,:3])
```

## Distancia ao centroide

```python
D1 = np.sqrt(np.sum((X-m1[:, np.newaxis])**2, axis = 0)) ### Distancia de todos os pontos ao centro
D2 = np.sqrt(np.sum((X-m2[:, np.newaxis])**2, axis = 0))
D3 = np.sqrt(np.sum((X-m3[:, np.newaxis])**2, axis = 0))
```

```python
D = np.vstack((D1, D2, D3))
ye = np.argmin(D, axis = 0)+1
```

```python
err = np.sum(y!=ye)/3000 # taxa de erro comparando y estimado e y real
acerto = 1-err
print('err', err , 'acerto ' , acerto)
```

```python
from sklearn.metrics import confusion_matrix
MC = confusion_matrix(y,ye)
MC
```

## Distancia de Mahalanobis

```python
S1 = np.cov(X[:,y==1])
print(S1)
```

```python
#valores teoricos dos acetatos

#m1 = np.array([-2,2])
#m2 = np.array([2,-2])
#m1 = np.array([0,0])
S1 = np.eye(2)* 1/4
S2 = np.array([[1/4,0],[0,1/4]])
S3 = np.array([[3.,0],[0,1/10]])
```

```python
from numpy.linalg import pinv
S1i = pinv(S1)
S2i = pinv(S2)
S3i = pinv(S3)
```

```python
D1 = np.sqrt(np.sum((X-m1[:,np.newaxis])*np.dot(S1i,X-m1[:,np.newaxis]), axis=0))
D2 = np.sqrt(np.sum((X-m2[:,np.newaxis])*np.dot(S2i,X-m2[:,np.newaxis]), axis=0))
D3 = np.sqrt(np.sum((X-m3[:,np.newaxis])*np.dot(S3i,X-m3[:,np.newaxis]), axis=0))
D = np.vstack ((D1,D2,D3))
print(D.shape)

ye = np.argmin(D, axis=0)+1

print(ye.shape , y.shape)

MC = confusion_matrix(y,ye)
print(MC)
print(np.sum(y!=ye))
```

```python
D = spd.squareform(spd.pdist(X.T),'mahalanobis') # pdist calcula a distancia entre todos os pontos. 
print(D.shape) #apenas √© relevante a distancia ao centro da classe

ye = np.argmin(D, axis=0)+1

print(ye.shape , y.shape)

MC = confusion_matrix(y,ye)
print(MC)
print(np.sum(y!=ye))
```

```python
plt.plot(X[0,y==1], X[1,y==1],'.')
plt.plot(X[0,y==2], X[1,y==2],'.')
plt.plot(X[0,y==3], X[1,y==3],'.')

plt.plot(X[0,y!=ye], X[1,y!=ye], 'ok', alpha=.2) #ver os erros
plt.axis('scaled')
```

## Classificador de distancia ao centroide com dados reais : IRIS

```python
from sklearn import datasets as DT

D = DT.load_iris()
X = D.data.T
y = D.target
```

```python
m0 = np.mean(X[:, y==0], axis = 1) # M√©dia de cada classe do iris dataset
m1 = np.mean(X[:, y==1], axis = 1)
m2 = np.mean(X[:, y==2], axis = 1)


D0 = np.sqrt(np.sum((X-m0[:,np.newaxis])**2, axis=0)) #calcular distancia de cada ponto ao centroide com a distancia euclidiana
D1 = np.sqrt(np.sum((X-m1[:,np.newaxis])**2, axis=0))
D2 = np.sqrt(np.sum((X-m2[:,np.newaxis])**2, axis=0))

D = np.vstack((D0,D1,D2))
ye= np.argmin(D, axis=0)

```

```python
MC = confusion_matrix(y,ye)
print(MC)
print(np.sum(y!=ye))
plt.plot(y)
plt.plot(ye)
```

## Usar a distancia de Mahalanopis como metrica de distancia

```python
m0 = np.mean(X1[:,y1==0], axis=1)
m1 = np.mean(X1[:,y1==1], axis=1)
m2 = np.mean(X1[:,y1==2], axis=1)
m3 = np.mean(X1[:,y1==3], axis=1)
m4 = np.mean(X1[:,y1==4], axis=1)
m5 = np.mean(X1[:,y1==5], axis=1)
m6 = np.mean(X1[:,y1==6], axis=1)
m7 = np.mean(X1[:,y1==7], axis=1)
m8 = np.mean(X1[:,y1==8], axis=1)
m9 = np.mean(X1[:,y1==9], axis=1)
```

```python
S0 = np.cov(X1[:,y1==0])
S1 = np.cov(X1[:,y1==1])
S2 = np.cov(X1[:,y1==2])
S3 = np.cov(X1[:,y1==3])
S4 = np.cov(X1[:,y1==4])
S5 = np.cov(X1[:,y1==5])
S6 = np.cov(X1[:,y1==6])
S7 = np.cov(X1[:,y1==7])
S8 = np.cov(X1[:,y1==8])
S9 = np.cov(X1[:,y1==9])
```

```python
from scipy.linalg import inv, pinv

S0i = pinv(S0) # inverso da matriz de cov
S1i = pinv(S1)
S2i = pinv(S2)
S3i = pinv(S3)
S4i = pinv(S4)
S5i = pinv(S5)
S6i = pinv(S6)
S7i = pinv(S7)
S8i = pinv(S8)
S9i = pinv(S9)
```

```python
Xtmp = X1 - m0[:,np.newaxis]
D0 = np.sqrt(np.sum(Xtmp*np.dot(S0i, Xtmp) , axis=0))

Xtmp = X1 - m1[:,np.newaxis]
D1 = np.sqrt(np.sum(Xtmp*np.dot(S1i, Xtmp) , axis=0))

Xtmp = X1 - m2[:,np.newaxis]
D2 = np.sqrt(np.sum(Xtmp*np.dot(S2i, Xtmp) , axis=0))

Xtmp = X1 - m3[:,np.newaxis]
D3 = np.sqrt(np.sum(Xtmp*np.dot(S3i, Xtmp) , axis=0))

Xtmp = X1 - m4[:,np.newaxis]
D4 = np.sqrt(np.sum(Xtmp*np.dot(S4i, Xtmp) , axis=0))

Xtmp = X1 - m5[:,np.newaxis]
D5 = np.sqrt(np.sum(Xtmp*np.dot(S5i, Xtmp) , axis=0))

Xtmp = X1 - m6[:,np.newaxis]
D6 = np.sqrt(np.sum(Xtmp*np.dot(S6i, Xtmp) , axis=0))

Xtmp = X1 - m7[:,np.newaxis]
D7 = np.sqrt(np.sum(Xtmp*np.dot(S7i, Xtmp) , axis=0))

Xtmp = X1 - m8[:,np.newaxis]
D8 = np.sqrt(np.sum(Xtmp*np.dot(S8i, Xtmp) , axis=0))

Xtmp = X1 - m9[:,np.newaxis]
D9 = np.sqrt(np.sum(Xtmp*np.dot(S9i, Xtmp) , axis=0))
```

```python
plt.plot(D0)
```

```python
D = np.vstack((D0, D1, D2, D3,D4,D5,D6,D7,D8,D9))
y1e = np.argmin(D, axis=0)

MC= confusion_matrix(y1,y1e)
print(MC) 
print(np.sum(y1!=y1e))
```

```python
## testar com o conjunto de teste


Xtmp = X2 - m0[:,np.newaxis]
D0 = np.sqrt(np.sum(Xtmp*np.dot(S0i, Xtmp) , axis=0))

Xtmp = X2 - m1[:,np.newaxis]
D1 = np.sqrt(np.sum(Xtmp*np.dot(S1i, Xtmp) , axis=0))

Xtmp = X2 - m2[:,np.newaxis]
D2 = np.sqrt(np.sum(Xtmp*np.dot(S2i, Xtmp) , axis=0))

Xtmp = X2 - m3[:,np.newaxis]
D3 = np.sqrt(np.sum(Xtmp*np.dot(S3i, Xtmp) , axis=0))

Xtmp = X2 - m4[:,np.newaxis]
D4 = np.sqrt(np.sum(Xtmp*np.dot(S4i, Xtmp) , axis=0))

Xtmp = X2 - m5[:,np.newaxis]
D5 = np.sqrt(np.sum(Xtmp*np.dot(S5i, Xtmp) , axis=0))

Xtmp = X2 - m6[:,np.newaxis]
D6 = np.sqrt(np.sum(Xtmp*np.dot(S6i, Xtmp) , axis=0))

Xtmp = X2 - m7[:,np.newaxis]
D7 = np.sqrt(np.sum(Xtmp*np.dot(S7i, Xtmp) , axis=0))

Xtmp = X2 - m8[:,np.newaxis]
D8 = np.sqrt(np.sum(Xtmp*np.dot(S8i, Xtmp) , axis=0))

Xtmp = X2 - m9[:,np.newaxis]
D9 = np.sqrt(np.sum(Xtmp*np.dot(S9i, Xtmp) , axis=0))

D = np.vstack((D0, D1, D2, D3,D4,D5,D6,D7,D8,D9))
y2e = np.argmin(D, axis=0)

MC= confusion_matrix(y2,y2e)
print(MC) 
print(np.sum(y2!=y2e))
```
